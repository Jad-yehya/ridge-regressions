```python
import pandas as pd
import numpy as np
# Import tqdm notebook
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science', 'ieee'])
import models
```

# Starting without categorical features

```python
df = pd.read_csv('dataset.csv', index_col=0)
df.head()
```

```python
# Transforming track_genre column into categorical
df['track_genre'] = df['track_genre'].astype('category')
df['key'] = df['key'].astype('category')
df['time_signature'] = df['time_signature'].astype('category')
# Transforming mode into boolean
df['mode'] = df['mode'].astype('bool')
```

```python
# Preparing data
df_num = df.select_dtypes(include=['int64', 'float64'])
df_num = df_num.drop(columns=['popularity'])
df_num = (df_num - df_num.mean()) / df_num.std()
y = df['popularity'].values.reshape(-1)
X = df_num.values
```

```python
# Adding the bias
n_X = []
t = np.array([1])
for i in range(len(X)):
    p = np.append(t, X[i])
    n_X.append(p)
X = np.array(n_X)
```

```python
# Splitting data into train and test
n_samples = X.shape[0]
p = np.random.permutation(n_samples)
X_train = X[p[:int(0.8*n_samples)]]
y_train = y[p[:int(0.8*n_samples)]]
X_test = X[p[int(0.8*n_samples):]]
y_test = y[p[int(0.8*n_samples):]]
```

## Starting the hyperparameter tuning

```python
GRID = np.linspace(0.5, 15, 30)
errors = []
for gamma in tqdm(GRID):
    model = models.GaussianKernelRidgeRegression(1, gamma, shape=X.shape[1])
    errors.append(np.mean(model.cross_validation(X_train[:1000], y_train[:1000], k=5)))
```

```python
plt.plot(GRID, errors)
plt.xlabel('gamma')
plt.ylabel('error')
plt.title('Evolution of the error with gamma')
plt.savefig('error_gamma.eps', format='eps')
```

```python
np.linspace(0.5, 15, 30)[np.argmin(errors)]
```

```python
# Create a grid of lambda 
GRID = np.linspace(0.01, 10, 20)
errors = []
for lambda_ in tqdm(GRID):
    model = models.GaussianKernelRidgeRegression(lambda_, 10.5, shape=X.shape[1])
    errors.append(np.mean(model.cross_validation(X[:1000], y[:1000], k=5)))
    
# Plot the error as a function of lambda
plt.plot(GRID, errors)
plt.xlabel('lambda')
plt.ylabel('error')
plt.title('Evolution of the error with lambda')
plt.savefig('error_lambda.eps', format='eps')
```

```python
# Create a grid of lambda 
GRID = np.linspace(0.01, 2.5, 20)
errors = []
for lambda_ in tqdm(GRID):
    model = models.GaussianKernelRidgeRegression(lambda_, 1, shape=X.shape[1])
    errors.append(np.mean(model.cross_validation(X[:1000], y[:1000], k=5)))

# Plot the error as a function of lambda
plt.plot(GRID, errors)
plt.xlabel('lambda')
plt.ylabel('error')
plt.title('Evolution of the error with lambda')
plt.savefig('error_lambda_precise.eps', format='eps')
```

```python
gaussian_model = models.GaussianKernelRidgeRegression(0.68, 10.5, shape=X.shape[1])
print('Training...')
gaussian_model.train(X_train[:2000], y_train[:2000])
```

```python
# scatter plot of the real and predicted values
plt.scatter(y_test[:100], gaussian_model.predict(X_test[:100]))
plt.xlabel('Real values')
plt.ylabel('Predicted values')
plt.title('Kernel Ridge Regression')
# plt.savefig('real_vs_predicted_kernel.eps', format='eps')
```

```python
y_pred = gaussian_model.predict(X_test[:200])
```

```python
print('MSE: ', np.mean((y_pred - y_test[:1000])**2))
print('MAE: ', np.mean(np.abs(y_pred - y_test[:1000])))
```

# With categorical features

```python
# Category columns:
cat = df.select_dtypes(include=['category', 'bool'])
df_full = pd.concat([df_num, cat], axis=1)
df_full.info()
```

Using one hot encoding for the categorical variables

```python
df_full = pd.get_dummies(df_full, drop_first=True)
```

```python
# Transform the boolean columns into 0 and 1
df_full['explicit'] = df_full['explicit'].astype(int)
df_full['mode'] = df_full['mode'].astype(int)
df_full.info()
```

```python
# Doing the same steps as before
y = df['popularity'].values.reshape(-1)
X = df_full.values
```

```python
n_X = []
t = np.array([1])
for i in range(len(X)):
    p = np.append(t, X[i])
    n_X.append(p)

X = np.array(n_X)
```

```python
# Splitting data into train and test
n_samples = X.shape[0]
p = np.random.permutation(n_samples)
X_train = X[p[:int(0.8*n_samples)]]
y_train = y[p[:int(0.8*n_samples)]]
X_test = X[p[int(0.8*n_samples):]]
y_test = y[p[int(0.8*n_samples):]]
```

```python
KRR = models.GaussianKernelRidgeRegression(0.68, 10.5, shape=X.shape[1])
errors = KRR.cross_validation(X_train[:1000], y_train[:1000], k=5)
```

```python
print(f' The mean error is {np.mean(errors)}')
```
